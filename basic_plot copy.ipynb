**Barplot:** mencari rata-rata biaya bulanan unutk setiap metode pembayaran menggunakan pandas


```python
df_bar= df.groupby('PaymentMethod').sum().reset_index()
df_bar
```

```python
df_bar.plot(x='PaymentMethod',y='MonthlyCharges',kind='bar')
```

**Barplot:** mencari rata-rata biaya bulanan unutk setiap metode pembayaran menggunakan matplotlib

```python
plt.bar(x='PaymentMethod',height='MonthlyCharges', data = df_bar)
plt.xticks(rotation = 90) #melakukan rotasi x_label nya 90 derajat
```

**Barplot:** mencari rata-rata biaya bulanan unutk setiap metode pembayaran menggunakan seaborn


```python
sns.barplot(x='PaymentMethod',y='MonthlyCharges', data = df_bar)
plt.xticks(rotation = 45) #melakukan rotasi x_label nya 45 derajat
```

**Multi-barplot :** mencari rata-rata biaya bulanan unutk setiap metode pembayaran dan berdasarkan churn menggunakan seaborn

```python
sns.barplot(x='MonthlyCharges',y='PaymentMethod', data=df, estimator=sum, hue='Churn')
```

**Stacked Multi-Bar Plot**

```python
df_multi_bar = df.groupby(['PaymentMethod','Churn'])['MonthlyCharges'].sum().reset_index()
df_multi_bar
```

```python
df_pivot_multibar = df_multi_bar.pivot_table(index='PaymentMethod',columns='Churn',values='MonthlyCharges')
df_pivot_multibar
```

```python
df_pivot_multibar.plot(kind='barh',stacked=False)
```
